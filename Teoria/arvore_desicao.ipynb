# Árvore de Regressão

A árvore de regressão é um modelo preditivo utilizado para prever um valor contínuo, como preços ou medidas, ao invés de classes discretas (como no caso das árvores de decisão para classificação). Sua lógica é baseada em particionar o conjunto de dados de entrada em subconjuntos cada vez menores, de forma a melhorar a precisão da previsão em cada subconjunto. Este processo é repetido até que os subconjuntos sejam homogêneos o suficiente, ou até que seja atingido algum critério de parada (como uma profundidade máxima).

## Funcionamento da Árvore de Regressão

A ideia principal de uma árvore de regressão é particionar os dados repetidamente, escolhendo as divisões (ou "splits") que resultam na menor soma dos erros quadráticos médios (MSE, ou "mean squared error"). O objetivo é encontrar uma maneira de dividir os dados que minimize o erro em cada nó.

### Etapas Básicas:

#### 1. Escolha de uma variável e um ponto de divisão (split):
- O modelo avalia todas as possíveis divisões em todas as variáveis preditoras (features) para encontrar o ponto que resulta na maior redução do erro.
- O algoritmo seleciona a variável que, quando dividida em dois grupos (à esquerda e à direita do ponto de divisão), resulta na menor soma dos erros quadráticos médios em cada grupo.

#### 2. Cálculo do valor de previsão em cada nó:
- Em cada nó da árvore, o valor previsto para os dados é a média dos valores da variável dependente (target) presentes naquele nó. Por exemplo, se estamos prevendo preços de casas, a previsão em cada nó será a média dos preços das casas que pertencem a esse nó.

#### 3. Critério de divisão:
- O critério de divisão é baseado no cálculo do erro quadrático médio (MSE). Para cada possível divisão de uma variável $x_j$ em um ponto de divisão $s$, o MSE é calculado como a soma ponderada dos erros quadráticos nos dois subgrupos criados pela divisão:
  
$$MSE(x_j, s) = \frac{N_L}{N} \cdot MSE_L + \frac{N_R}{N} \cdot MSE_R$$

  Onde:
  - $N$ é o número total de amostras,
  - $N_L$ e $N_R$ são os números de amostras no grupo da esquerda e da direita, respectivamente,
  - $MSE_L$ e $MSE_R$ são os erros quadráticos médios no grupo da esquerda e da direita, respectivamente.

  - Também temos o grau de pureza de uma arvore.

#### 4. Recursão:
- A árvore continua particionando os dados recursivamente, aplicando a mesma lógica em cada subgrupo de dados resultante. A recursão continua até que uma condição de parada seja atingida, como:
  - Um número mínimo de amostras em um nó,
  - Uma profundidade máxima da árvore,
  - A redução do erro em uma divisão ser pequena demais para justificar a continuação.

#### 5. Previsão:
- Depois de a árvore ser construída, para fazer uma previsão para uma nova amostra, percorremos a árvore a partir da raiz, seguindo as condições de divisão nos nós até atingir um nó folha. O valor da previsão é a média dos valores da variável dependente das amostras presentes naquela folha.

## Exemplo de Árvore de Regressão

Vamos considerar um exemplo simples de um conjunto de dados com duas variáveis, $x_1$ e $x_2$, e a variável alvo $y$, que queremos prever.

|  x_1  |  x_2  |  y  | 
|-------|-------|-----|
|  1.0  |  3.0  | 2.0 |
|  2.0  |  2.0  | 4.0 |
|  3.0  |  4.0  | 5.0 |
|  4.0  |  2.5  | 6.0 |
|  5.0  |  5.0  | 8.0 |

O algoritmo começaria testando várias divisões possíveis. Digamos que ele primeiro avalia dividir a variável $x_1$ em um ponto $s = 3.5$, separando os dados em dois grupos:

- **Grupo 1 (esquerda):** $x_1 \leq 3.5 \rightarrow$ { (1.0, 3.0, 2.0), (2.0, 2.0, 4.0), (3.0, 4.0, 5.0) }
- **Grupo 2 (direita):** $x_1 > 3.5 \rightarrow$ { (4.0, 2.5, 6.0), (5.0, 5.0, 8.0) }

Agora o modelo calcularia o MSE para esses dois grupos. O erro é calculado como:

$$MSE_L = \frac{1}{3} \sum_{i=1}^{3} (y_i - \bar{y}_L)^2$$

$$
MSE_R = \frac{1}{2} \sum_{i=4}^{5} (y_i - \bar{y}_R)^2
$$

Onde $\bar{y}_L$ e $\bar{y}_R$ são as médias dos valores de $y$ nos dois grupos.

Se essa divisão resulta na maior redução do MSE comparada com outras possíveis divisões, o algoritmo a seleciona. O processo então continua recursivamente para os dois subgrupos.

## Aspectos Matemáticos

O ponto principal da árvore de regressão é a escolha dos pontos de divisão que minimizam o erro quadrático médio. Formalmente, a função de erro a ser minimizada em cada nó pode ser expressa como:

$$Custo = \frac{1}{N} \sum_{i=1}^{N} (y_i - \hat{y}_i)^2$$

Onde $\hat{y}_i$ é a média dos valores $y_i$ no grupo atual.

O algoritmo utiliza uma abordagem gulosa ("greedy"), selecionando a divisão que resulta na maior redução do erro, sem considerar futuras divisões.

## Vantagens e Desvantagens

### Vantagens:
- **Interpretação fácil:** Árvores de regressão são bastante intuitivas e fáceis de visualizar.
- **Capacidade de lidar com não-linearidades:** Árvores são capazes de capturar relações não-lineares entre variáveis, pois realizam divisões nos dados sem assumir uma forma paramétrica subjacente.
- **Lida com interações entre variáveis:** Árvores de decisão podem facilmente lidar com interações complexas entre as variáveis preditoras.

### Desvantagens:
- **Propenso a overfitting:** Árvores podem se ajustar demais aos dados de treinamento se não forem adequadamente controladas (por exemplo, usando poda).
- **Descontinuidades:** As previsões feitas por uma árvore de regressão podem ser descontinuas, o que pode não ser desejável em certos contextos.
- **Estabilidade:** Árvores de regressão podem ser instáveis, no sentido de que pequenas mudanças nos dados podem resultar em uma árvore bastante diferente.

## Melhoria do Modelo: Random Forests

Um dos principais problemas das árvores de decisão é a tendência de superajuste (overfitting) aos dados de treinamento. Uma solução para isso é o **Random Forest**, que é um conjunto ("ensemble") de múltiplas árvores de regressão construídas a partir de diferentes amostras dos dados. A previsão final é feita por uma média das previsões de todas as árvores, o que geralmente melhora a precisão e reduz o overfitting.

````python
# Definindo cores
BLACK = "\033[30m"
RED = "\033[31m"
GREEN = "\033[32m"
YELLOW = "\033[33m"
BLUE = "\033[34m"
MAGENTA = "\033[35m"
CYAN = "\033[36m"
WHITE = "\033[37m"
RESET = "\033[0m"  # Reseta a cor

print(f"{RED}Texto em vermelho{RESET}")
print(f"{GREEN}Texto em verde{RESET}")
print(f"{YELLOW}Texto em amarelo{RESET}")
print(f"{BLUE}Texto em azul{RESET}")
print(f"{MAGENTA}Texto em magenta{RESET}")
print(f"{CYAN}Texto em ciano{RESET}")
print(f"{WHITE}Texto em branco{RESET}")
````

````python
import numpy as np
#Conjunto de dados.
# Conjunto de dados
data = np.array([
    [50, 1, 1990, 150],
    [60, 2, 1985, 180],
    [70, 2, 2000, 210],
    [80, 3, 2015, 240],
    [90, 3, 2010, 270],
    [100, 4, 2020, 300],
    [110, 4, 2005, 330],
    [120, 5, 2018, 360]
])

# Separando features e target
X = data[:, :3] # features (X1, X2, X3)
y = data[:, 3]
````

````python
def	mean_squared_error(y_true, y_pred):
	return np.mean((y_true - y_pred) ** 2)
````

````python
# Funcao para encontrar a melhor divisao
def	best_split(X, y):
	best_mse = float('inf')
	best_feature = None
	best_threshold = None

	for feature_index in range(X.shape[1]):
		thresholds = np.unique(X[:, feature_index])
		print(f"{GREEN}Thresholds: {thresholds}{RESET}")

		for threshold in thresholds:
			left_mask = X[:, feature_index] <= threshold
			rigth_mask = X[:, feature_index] > threshold
			print(f"{GREEN}For each Threshold : {threshold}\n{RESET}{YELLOW}Left mask: {left_mask} Rigth mask: {rigth_mask}{RESET}")

			if np.any(left_mask) and np.any(rigth_mask):
				left_y = y[left_mask]
				rigth_y = y[rigth_mask]
				print(f"{YELLOW}Left Y:{left_y} Rigth Y: {rigth_y}{RESET}")

				left_mean = np.mean(left_y)
				rigth_mean = np.mean(rigth_y)
				print(f"{BLUE}Left mean: {left_mean} Rigth mean: {rigth_mean}{RESET}")

				mse_left = mean_squared_error(left_y, np.full(left_y.shape, left_mean))
				mse_rigth = mean_squared_error(rigth_y, np.full(rigth_y.shape, rigth_mean))
				print(f"{BLUE}MSE Left: {mse_left} MSE Rigth: {mse_rigth}{RESET}")

				# Média ponderada MSE
				total_mse = ((len(left_y) * mse_left) + (len(rigth_y) * mse_rigth)) / len(y)
				print(f"{MAGENTA}Total MSE: {total_mse}{RESET}")

				if total_mse < best_mse:
					best_mse = total_mse
					best_feature = feature_index
					best_threshold = threshold
					print(f"{MAGENTA}Best Mse: {best_mse}\nBest Feature: {best_feature}\nBest Threshold: {best_threshold}\n{RESET}")
				print(f"{CYAN}Best Mse: {best_mse}{RESET}")

	print(f"Melhor feature: {best_feature}, Melhor threshold: {best_threshold}")
	return best_feature, best_threshold

feature, threshold = best_split(X, y)
#print(f"Melhor feature: {feature}, Melhor threshold: {threshold}")
````

````python
# Função para construir a árvore de forma recursiva
def build_tree(X, y, depth=0, max_depth=5, min_samples_split=2):
	num_samples, num_features = X.shape

	# Critérios de parada: profundidade máxima ou número mínimo de amostras
	if depth >= max_depth or num_samples < min_samples_split:
		leaf_value = np.mean(y)
		return {"leaf": True, "value": leaf_value}

	# Encontrando a melhor divisão
	best_feature, best_threshold = best_split(X, y)

	# Se não houver divisão válida, retorna o valor da média do nó (folha)
	if best_feature is None:
		leaf_value = np.mean(y)
		return {"leaf": True, "value": leaf_value}

	# Separando os dados em left e right com base no best_feature e best_threshold
	left_mask = X[:, best_feature] <= best_threshold
	right_mask = X[:, best_feature] > best_threshold
	left_X, left_y = X[left_mask], y[left_mask]
	right_X, right_y = X[right_mask], y[right_mask]

	# Recursivamente construindo as subárvores
	left_subtree = build_tree(left_X, left_y, depth + 1, max_depth, min_samples_split)
	right_subtree = build_tree(right_X, right_y, depth + 1, max_depth, min_samples_split)

	# Retornando a árvore com a melhor divisão encontrada
	return {
		"leaf": False,
		"feature": best_feature,
		"threshold": best_threshold,
		"left": left_subtree,
		"right": right_subtree
	}

# Exemplo de uso:
tree = build_tree(X, y)
print(tree)
````

````python
def print_tree(node, depth=0, side=""):
    # Define o espaço para organizar visualmente a estrutura
    indent = " " * (8 * depth)
    
    # Se o nó for uma folha, imprime o valor da previsão (média)
    if node["leaf"]:
        print(f"{indent}{side}[Preço médio: {node['value']:.2f}]")
    else:
        # Nome da feature e threshold para a condição do nó
        feature_name = f"Feature {node['feature']}"
        
        # Imprime a condição da divisão atual
        print(f"{indent}{side}[{feature_name} ≤ {node['threshold']:.2f}]")
        
        # Imprime o ramo da esquerda (Sim) com uma nova indentação e profundidade aumentada
        print(f"{indent}    ├── Sim:")
        print_tree(node["left"], depth + 1, side="  ")
        
        # Imprime o ramo da direita (Não) com a mesma profundidade
        print(f"{indent}    └── Não:")
        print_tree(node["right"], depth + 1, side="  ")

# Exemplo de uso com a árvore gerada:
#tree = build_tree(X, y)
print_tree(tree)
````
